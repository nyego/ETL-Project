```python
import pandas as pd
from sqlalchemy import create_engine

import json

import numpy as np

import pymysql
pymysql.install_as_MySQLdb()

#from config import mypass
```

```python
#Store CSV into a Data Frame
#Convert Zip Code (float into string)
csv_file = "Raw_Data/311_call_center.csv"
call_center_df = pd.read_csv(csv_file, dtype={"Zip Code":str}, keep_default_na=False)
call_center_df.head()
```

```python
#Drop all NaN values
call_center_df = call_center_df.dropna(how='any')
call_center_df.head()
```

```python
#Remove unnecessary columns
new_call_center_df = call_center_df[["Time", "Department Name", "Service Name", "Zip Code"]].copy()
new_call_center_df.head()
```

```python
#Rename columns
new_call_center_df = new_call_center_df.rename(columns=\
                                      {"Time": "Time_",\
                                      "Department Name": "Department_Name",\
                                      "Service Name": "Service_Name",\
                                      "Zip Code": "Zip_Code"})
new_call_center_df.head()
```

```python
#Truncate data to make it easier to convert to database
new_call_center_trunc=new_call_center_df.truncate(before=None, after=499)
new_call_center_trunc
```

```python
new_call_center_trunc.dtypes
```

```python
#Store json into a Data Frame
data = json.load(open('Raw_Data/Project.json'))

census_df = pd.DataFrame(data["data"])
census_df.head() 
```

```python
#Remove unnecessary columns
new_census_df = census_df[[8, 9, 10, 11, 12, 13, 14]].copy()
new_census_df.head()
```

```python
#Rename columns
new_census_df = new_census_df.rename(columns=\
                                        {8: "Zip_Code",\
                                        9: "Total_Population",\
                                        10: "Median_Age", \
                                        11: "Total_Males",\
                                        12: "Total_Females",\
                                        13: "Total_Household",\
                                        14: "Average_Household_Size"})
new_census_df.head()
```

```python
new_census_df.dtypes
```

```python
#Connect to local database
rds_connection_string = "root:Taurai3055%@127.0.0.1/advertising_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
#Confirm Tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
new_call_center_trunc.to_sql(name='call_center', con=engine, if_exists='replace', index=True)
```

```python
#Use pandas to load json converted DataFrame into database
new_census_df.to_sql(name='census', con=engine, if_exists='replace', index=True)
```
